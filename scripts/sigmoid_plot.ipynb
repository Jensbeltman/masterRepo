```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

```python
def sig(x,g,c):
    return 1-(1/(1+exp(-g*(x-c))))

x = np.arange(0,10,0.01)
g = np.arange(1,5,1)
y = [[sig(xv,gv,5) for xv in x ] for gv in g]
```

```python
for r in y:
    plt.plot(x,r)

plt.legend(["g = "+str(gv) for gv in g])
plt.xlabel("x")
plt.ylabel("$\sigma (x)$")
plt.title("")
plt.tight_layout()
plt.savefig('report/sigmoid.eps', format='eps')
plt.show()
```
