```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

```python
def sig(x,g,c):
    return 1/(1+exp(-g*(x-c)))

x = np.arange(-3,3,0.1)
g = np.arange(1,10,2)
y = [[sig(xv,gv,0) for xv in x ] for gv in g]
```

```python
for r in y:
    plt.plot(x,r)

plt.legend(["$\sigma$("+str(gv)+")" for gv in g])
plt.show()
```
